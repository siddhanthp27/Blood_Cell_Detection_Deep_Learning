```python
import matplotlib.pyplot as plt
%matplotlib inline
from PIL import Image
import numpy as np
import pickle
```

```python
train = open("train.lst", "r")
train_list = train.readlines()
```

```python
folder = './pkl_files/'
```

```python
for iter in range(len(train_list)):
    l = train_list[iter]
    l = l.split('\t')

    idx, width, heigth = l[0], l[3], l[4]
    name = l[-1]

    idx = int(idx)
    width = int(float(width))
    heigth = int(float(heigth))

    l = l[:-1]
    l = l[5:]

    name = name.split('\n')

    name = name[0]

    # img = Image.open(name)

    # plt.imshow(img)

    area = np.full((480, 640), fill_value=5)

    for i in range(len(l) / 5):
        class_id = l[5*i]
        x_min = float(l[5*i+1]) * width
        y_min = float(l[5*i+2]) * heigth
        x_max = float(l[5*i+3]) * width
        y_max = float(l[5*i+4]) * heigth
        x_min = int(x_min)
        y_min = int(y_min)
        x_max = int(x_max)
        y_max = int(y_max)
#         print(class_id)
        for j in range(y_min, y_max):
            for k in range(x_min, x_max):
                area[j][k] = class_id

    pickle.dump(area, open(folder + 'area_' + name.split('/')[-1] + '.pkl', 'wb'))
    # plt.imshow(area)
#         fig, ax = plt.subplots()

#         im = ax.imshow(area, cmap='jet')

#         fig.colorbar(im, ax=ax)

#         plt.savefig('area' + name.split('/')[-1] + '.png')
#         im.save('area' + name.split('/')[-1])
```
